### analysis column `Salary Data`

```python
pd.set_option('display.max_colwidth', None)
```

```python
df['Salary Data']
```

##### make a copy of the dataframe to avoid changes to the main dataframe

```python
new_df = df['Salary Data'].copy()
new_df
```

##### prints all the keys in the `Salary Data` column

```python
keys = set()

for salary in df['Salary Data']:
    try:
        temp = json.loads(salary)  
        keys.update(temp.keys()) 
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")

print('All keys in the Salary Data:\n', keys)
```

#### prints all nor-null values in the column

```python
def non_null_data(salary_data):
    salary_data = json.loads(salary_data)
    return any(value is not None for value in salary_data.values())

df_filtered = new_df[new_df.apply(non_null_data)]
df_filtered
```

#### Splitting value `Salary Data` into Columns

```python
df_split_salary = pd.json_normalize(new_df.apply(json.loads))
df_split_salary
```

####  statistics the `Salary Data` column 

```python
null_values = df_split_salary.isnull().sum().sum()
not_null_values = df_split_salary.notnull().sum().sum()
total_data = df_split_salary.size  
total_missing = df_split_salary.isnull().sum().sum()

percentage = (total_missing / total_data) * 100

print(f'Null values in data\t: {null_values}')
print(f'Not null values in data\t: {not_null_values}')
print(f'Dimension of Data\t: {total_data}')
print(f'Percentage missing values : {percentage:.2f}%')
```

##### Because the missing values are very high, I decided to delete the column `Salary data`
